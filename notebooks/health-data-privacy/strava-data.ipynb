```python
import gpxpy
import pandas as pd

'''
Inspired by gpxpy post here: https://www.gpxz.io/blog/gpx-file-to-pandas
Data from https://academictorrents.com/details/af55533bf8229c3bff260b77a652f8b8058f6c9e
'''

# Load gpx.
gpx_path = 'data/Sport2/Athlete1/1.gpx'
with open(gpx_path) as f:
    gpx = gpxpy.parse(f)

# for filename in os.listdir('data/Sport2/Athlete1'):
points = []
for segment in gpx.tracks[0].segments:
    for p in segment.points:
        points.append({
            'time': p.time,
            'latitude': p.latitude,
            'longitude': p.longitude,
            'elevation': p.elevation,
        })
df = pd.DataFrame.from_records(points)
df
```
